# Passo a Passo: Detecção de Gatos
Este guia irá explicar como configurar, treinar e avaliar um modelo de detecção de gatos usando um dataset de imagens de gatos e cachorros.

## Passo 1: Importar as Bibliotecas Necessárias

Primeiro, vamos importar as bibliotecas necessárias. Isso inclui TensorFlow para construir e treinar o modelo, e Matplotlib para visualização.


```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
import matplotlib.pyplot as plt
```

## Passo 2: Preparar os Dados

Vamos carregar e preparar o dataset "Cats vs Dogs".


```python
import tensorflow_datasets as tfds

# Carregando o dataset Cats vs Dogs
(raw_train, raw_validation, raw_test), metadata = tfds.load(
    'cats_vs_dogs',
    split=['train[:80%]', 'train[80%:90%]', 'train[90%:]'],
    with_info=True,
    as_supervised=True,
)

# Função para redimensionar as imagens e normalizar os pixels
IMG_SIZE = 160

def format_example(image, label):
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
    image = image / 255.0
    return image, label

train = raw_train.map(format_example)
validation = raw_validation.map(format_example)
test = raw_test.map(format_example)

BATCH_SIZE = 32
SHUFFLE_BUFFER_SIZE = 1000

train_batches = train.shuffle(SHUFFLE_BUFFER_SIZE).batch(BATCH_SIZE)
validation_batches = validation.batch(BATCH_SIZE)
test_batches = test.batch(BATCH_SIZE)
```

## Passo 3: Construir o Modelo

Vamos construir um modelo de rede neural convolucional (CNN) para detecção de gatos.


```python
# Construindo o modelo CNN
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 3)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])
```

## Passo 4: Compilar o Modelo

Vamos compilar o modelo usando o otimizador Adam e a função de perda binária cruzada.


```python
# Compilando o modelo
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

## Passo 5: Treinar o Modelo

Vamos treinar o modelo com os dados de treinamento.


```python
# Treinando o modelo
history = model.fit(
    train_batches,
    epochs=10,
    validation_data=validation_batches
)
```

## Passo 6: Avaliar o Modelo

Após o treinamento, vamos avaliar o modelo com os dados de teste.


```python
# Avaliando o modelo
loss, accuracy = model.evaluate(test_batches)
print(f'Loss: {loss}, Accuracy: {accuracy}')
```

## Passo 7: Fazer Previsões e Visualizar

Vamos fazer previsões com algumas imagens de teste e visualizar os resultados.


```python
# Fazendo previsões
for image_batch, label_batch in test_batches.take(1):
    predictions = model.predict(image_batch)
    plt.figure(figsize=(10, 10))
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(image_batch[i])
        if predictions[i] > 0.5:
            plt.title("Gato")
        else:
            plt.title("Cachorro")
        plt.axis("off")
    plt.show()
```
